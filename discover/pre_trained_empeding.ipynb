```python
from navec import Navec
navec = Navec.load('/Volumes/UNTITLED/navec_hudlit_v1_12B_500K_300d_100q.tar')
navec.get('красиво')
```

```python
from gensim.models import Word2Vec, KeyedVectors

model = Word2Vec(vector_size=300)

keys = navec.vocab.words
vecs = [[item + 1 for item in navec.get(key)] for key in keys]

model.wv.add_vectors(keys, vecs)
```

```python
from random import choices

sentence = ['красота', ]
def next_word():
    similars = model.wv.most_similar(sentence, topn=50)
    words, weights = list(zip(*similars))
    return choices(words, weights)[0]

for _ in range(25):
    sentence.append(next_word())
print(sentence)
print()
```
